```python
from pandasql import sqldf
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
file_path = '../Data/Raw/participants.csv'
participants_df = pd.read_csv(file_path, header=0)
```

```python
participants_df
```

```python
participants_df = sqldf('''
                        SELECT
                            DISTINCT participant_id,
                            time_taken,
                            fluent_languages AS fluent_in,
                            age,
                            sex,
                            ethnicity_simplified AS ethnicity,
                            country_of_birth,
                            country_of_residence,
                            nationality
                        FROM participants_df
                        WHERE status = 'APPROVED'
                        ''')
```

```python
participants_df
```

```python
# Save the processed data to a new CSV file
participants_df.to_csv('../Data/Processed/participants.csv', index=False)
```
