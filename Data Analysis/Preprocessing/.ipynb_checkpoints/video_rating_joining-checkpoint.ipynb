```python
from pandasql import sqldf
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
file_path = '../Data/Processed/rating.csv'
rating_df = pd.read_csv(file_path, header=0, index_col=[0,1])
```

```python
rating_df
```

```python
# Read the CSV file into a DataFrame
file_path = '../Data/Processed/video.csv'
video_df = pd.read_csv(file_path, header=0, index_col=[9])
```

```python
video_df
```

```python
video_rating_df = sqldf('''
                        SELECT
                            rating_df.participant_id,
                            rating_df.video_id,
                            video_df.wanderSpeed AS wander_speed,
                            video_df.wanderRoundness AS wander_roundness,
                            video_df.wanderCycleRate AS wander_cycle_rate, 
                            video_df.blinkTemperature AS blink_temperature,
                            video_df.blinkSlope AS blink_slope,
                            video_df.blinkCycleRate AS blink_cycle_rate,
                            video_df.beepPitch AS beep_pitch,
                            video_df.beepSlope AS beep_slope,
                            video_df.beepCycleRate AS beep_cycle_rate,
                            rating_df.joy AS joy_intensity,
                            rating_df.sadness AS sadness_intensity,
                            rating_df.fear AS fear_intensity,
                            rating_df.anger AS anger_intensity,
                            rating_df.disgust AS disgust_intensity,
                            rating_df.surprise AS surprise_intensity,
                            rating_df.other_text AS other_emotion,
                            rating_df.other AS other_intensity,
                            rating_df.pleasure,
                            rating_df.arousal,
                            rating_df.dominance,
                            rating_df.appraisal
                        FROM rating_df 
                        JOIN video_df ON rating_df.video_id = video_df.video_id
                        ''')
```

```python
video_rating_df
```

```python
# Save the processed data to a new CSV file
video_rating_df.to_csv('../Data/Processed/video_rating.csv', index=False)
```
